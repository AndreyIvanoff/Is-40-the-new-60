```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np
```

```python
#http://vestnik.mednet.ru/content/view/166/30/lang,ru/
df_d = pd.read_csv("data/doctors.csv",sep='\t')
df_d.drop(columns=['мужчины','женщины'],inplace=True)
df_d.columns=('age_string','percent')
df_d['left'] = pd.Series([15,25,30,35,40,45,50,55,60,65,70])
df_d['right'] = pd.Series([24,29,34,39,44,49,54,59,64,69,100])
#df_d.drop([0,10],inplace=True)
```

```python
#https://insights.stackoverflow.com/survey/2019
#
df_p = pd.read_csv("data/programmers.csv",sep='\t',header=None)
df_p.columns=('age_string','percent')
df_p['left'] = pd.Series([10,15,20,25,30,35,40,45,50,55,60])
df_p['right'] = pd.Series([14,19,24,29,34,39,44,49,54,59,100])
#df_p.drop([0,10],inplace=True)
```

```python
#https://rosstat.gov.ru/storage/mediabank/Trud_2021.pdf
df_e = pd.read_csv("data/education.csv",sep='\t',header=None)
df_e.columns=('age_string','value')
max_value = 6676
df_e['percent'] = np.round(df_e['value']*100 / max_value,decimals=1)
df_e['left']  = pd.Series([15, 20, 30, 35, 40, 50, 60, 65, 70])
df_e['right'] = pd.Series([19, 29, 34, 39, 49, 59, 64, 69, 100])
#df_e.drop([8],inplace=True)
```

```python
df_info = pd.read_csv("data/connection_and_info.csv",sep='\t',header=None)
df_info.columns=('age_string','value')
max_value = 1366
df_info['percent'] = np.round(df_info['value']*100 / max_value,decimals=1)
df_info['left']  = pd.Series([15, 20, 30, 35, 40, 50, 60, 65, 70])
df_info['right'] = pd.Series([19, 29, 34, 39, 49, 59, 64, 69, 100])
```

```python
ymin = 14
ymax = 75

fig = plt.figure(figsize=(6,1.5), dpi=640);
gs = GridSpec(1, 2, width_ratios=[1, 1],hspace=0.0)
ax1 = fig.add_subplot(gs[0])

ax1.barh(
    y=df_info['left']+(df_info['right']-df_info['left'])/2,
    height=df_info['right']-df_info['left'],
    width=df_info['percent'],
    color='gray',
    alpha=0.5,
    label='Деятельность в области информации и связи **'
)

ax1.barh(
    y=df_p['left']+(df_p['right']-df_p['left'])/2,
    height=df_p['right']-df_p['left'],
    width=df_p['percent'],
    color='red',
    alpha=0.5,
    label='Профессиональные разработчики *'
)
#ax1.set_xscale('log')
#ax1.set_xlim([0,100])
ax1.set_xticks([])
ax1.set_ylim([ymin,ymax])
ax1.set_yticks(df_info['left'])
ax1.set_yticklabels(df_info['left'],fontsize=7)
ax1.yaxis.tick_right()
ax1.invert_yaxis()
ax1.invert_xaxis()
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(True)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.legend(
    loc='lower center',
    fontsize=6,
    bbox_to_anchor=(0.44, -0.3),
    reverse=True,
    frameon=False,
    markerfirst=False
)

ax2 = fig.add_subplot(gs[1])
ax2.barh(
    y=df_e['left']+(df_e['right']-df_e['left'])/2,
    height=(df_e['right']-df_e['left']),
    width=df_e['percent'],
    color='gray',
    alpha=0.5,
    label='Деятельность в области образования **'
)
ax2.barh(
    y=df_d['left']+(df_d['right']-df_d['left'])/2,
    height=df_d['right']-df_d['left'],
    width=df_d['percent'],
    color='black',
    alpha=0.5,
    label='Врачебные кадры ***'
)
#ax2.set_xscale('log')
#ax2.set_xlim([0,100])
ax2.set_xticks([])
ax2.set_ylim([ymin,ymax])
ax2.set_yticks(df_info['left'])
ax2.set_yticklabels([])
ax2.set_ylabel('возраст, лет',fontsize=6,labelpad=-1)
ax2.invert_yaxis()
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(True)
ax2.legend(loc='lower center',fontsize=6,bbox_to_anchor=(0.475, -0.3),frameon=False)

ax2.annotate('*** Леонов С. А. и др. Характеристика врачебных кадров\nразного профиля в субъектах Российской Федерации //\nСоциальные аспекты здоровья населения. – 2010. – Т. 13. – №. 1. – С. 4.',
             (0,0), (0, -25),
             xycoords='axes fraction',
             textcoords='offset points',
             va='top',fontsize=4)

ax1.annotate('* Developer Survey Results 2019, Stack Overflow\nhttps://insights.stackoverflow.com/survey/2019',
             (-0.1225,0), (-0.1225, -25),
             xycoords='axes fraction',
             textcoords='offset points',
             va='top',fontsize=4)

ax1.annotate('** Труд и занятость в России. 2021: Стат.сб./Росстат – M., 2021. – 177 c.',
             (-0.1225,0), (-0.1225, -35),
             xycoords='axes fraction',
             textcoords='offset points',
             va='top',fontsize=4)

fig.suptitle("Куда исчезают профессиональные разработчики после 30?",fontsize=7)
plt.savefig("over30.png", bbox_inches='tight')
```
